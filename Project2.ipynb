```python
query_url = "http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key=6f6f8b7af92a47298e124855191601&q=44113&format=json&date=2018-01-01&enddate=2018-12-31&includelocation=44113&tp=24"
print(query_url)
```

```python
weather_response = requests.get(query_url)
weathers = weather_response.json()
print(f"The weather API responded with: {weathers}.")
```

```python
pprint(weathers)
```

```python
date1 = weathers["data"]["weather"][0]['date']
temp = weathers["data"]["weather"][0]['hourly'][0]['tempF']
pprint(date1)
pprint(temp)
```

```python
Cle_date= []
Cle_temp = []
for r in weathers["data"]["weather"]:
    Cle_date.append(r["date"])
    Cle_temp.append(float(r['hourly'][0]['tempF']))
    
Cle_date
```

```python
d = {'Date':Cle_date, 'Temp':Cle_temp}
```

```python
Cle_df = pd.DataFrame(d)
Cle_df
```

```python
file = 'Data_Cle.csv'
csv_file = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
csv_file.head()
```

```python
merge_table = pd.merge(csv_file, Cle_df, on ='Date')
merge_table
```

```python
rename_df = merge_table.rename(columns={"Temp_x":"Actual Temp","Temp_y":"Forcast Temp"})
rename_df
```

```python
rename_df["Difference"] = rename_df['Actual Temp'] - rename_df['Forcast Temp']
```

```python
rename_df.to_csv("Final_result.csv", index=False, header=True)
```
